```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from google.colab import files
uploaded = files.upload()




df = pd.read_csv('ATfatalities.csv')
```

```python
# Step 3: Plot the timeline with dot size and color by fatalities
sns.set(style="whitegrid")
plt.figure(figsize=(14, 2.5))
sns.scatterplot(
    x='Year',
    y=[1] * len(df),  # All dots on a flat horizontal line
    size='Fatalities',
    sizes=(20, 800),
    hue='Fatalities',
    palette='Reds',
    data=df,
    legend=False,
    edgecolor='black'
)

# Step 4: Annotate major incidents
for i, row in df.iterrows():
    if row['Fatalities'] >= 150 or row['Year'] == 2025:
        plt.text(row['Year'], 1.05, f"{row['Year']} ({row['Fatalities']})", ha='center', fontsize=8)

# Step 5: Final formatting
plt.yticks([])
plt.xlabel('Year')
plt.title('Major Airliner Accidents in India â€” Dot Size and Color = Fatalities', fontsize=13)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
